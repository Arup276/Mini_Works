```python
categories = ['Cat','Dog']
dirt = "C:/Users/AArup/Desktop/365/52_Day(keras_catdog)/PetImages"

training_data = []
def create_train_data():
    for category in categories:
        join_path = os.path.join(dirt,category)
        class_num = categories.index(category)
        print(join_path)
        for imgs in os.listdir(join_path):
            print(imgs)
            try:
                img = cv2.imread(os.path.join(join_path,imgs))
                re_img = cv2.resize(img,(60,60))
                training_data.append([re_img,class_num])
            except Exception as e:
                print(e)
                
create_train_data()
```

```python
print(len(training_data))
```

```python
random.shuffle(training_data)
for i in training_data[:10]:
    print(i[1])
```

```python
# X and y is list but our model dos't accept list so we have to convert it into array
X = []
y = []
for image,label in training_data: 
    X.append(image) # image
    y.append(label) # label
X = np.array(X).reshape(-1,60,60,3) # -1 for how many feature we have and reshape for scale all images in a fix size
```
